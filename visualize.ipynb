```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
index_list = ['KoBERT', 'Lmkor-BERT', 'Lmkor-ALBERT', 'Lmkor-ELECTRA', 'ChatGPT3.5-turbo', 'ChatGPT4', 'HyperClova']

acc_dict = {
    'Movie' : [0.8463, 0.8805, 0.8545, 0.8845, 0.8381, 0.8863, 0.8156],
    'Game' : [0.7869, 0.828, 0.8005, 0.829, 0.7501, 0.7939, 0.7405],
    'Shopping' : [0.9162, 0.936, 0.9275, 0.94, 0.9218, 0.9411, 0.9153],
}

f1_dict = {
    'Movie' : [0.8453, 0.88, 0.8508, 0.8856, 0.8303, 0.8804, 0.8016],
    'Game' : [0.7859, 0.8277, 0.7949, 0.8292, 0.6869, 0.7543, 0.6661],
    'Shopping' : [0.9151, 0.9366, 0.9278, 0.9409, 0.9213, 0.9404, 0.9192],
}

for key in acc_dict.keys():
    acc_dict[key] = [round(x*100, 1) for x in acc_dict[key]]

for key in f1_dict.keys():
    f1_dict[key] = [round(x*100, 1) for x in f1_dict[key]]


acc_df = pd.DataFrame(acc_dict, index=index_list)
f1_df = pd.DataFrame(f1_dict, index=index_list)
total_df = pd.concat([acc_df, f1_df], axis = 1, keys=['Accuracy', 'F1'])
```

```python
total_df.T
```

```python
total_df.plot(kind='bar', figsize=(14, 5), width = 0.7, rot = 0)
plt.ylim(60, 95)
plt.show()
```
